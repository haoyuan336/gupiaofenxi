```python
print("Tensorflow Version : {}".format(tf.__version__))
code = "cn_000596"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data_csv = pd.read_csv("{}_data.csv".format(code))
```

```python
data_csv.head()

# plt.scatter(data_csv.num8, data_csv.num10)
```

```python
x = data_csv.iloc[:, 1: -1]
y = data_csv.iloc[:, -1]
modle = tf.keras.Sequential([
    tf.keras.layers.Dense(100,input_shape=(90,), activation="relu"),
    tf.keras.layers.Dense(10,input_shape=(10,), activation="relu"),
    tf.keras.layers.Dense(10,input_shape=(10,), activation="relu"),
    tf.keras.layers.Dense(1)
])
modle.summary()
```

```python
modle.compile(optimizer="adam", loss = "mse")
modle.fit(x,y,epochs=1000)
```

```python
test = data_csv.iloc[:10,1:-1]
# print(test)
print(y.iloc[:10,])
modle.predict(test)
```
